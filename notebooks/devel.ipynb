```python
import gradio as gr

def create_input_fields(processor, identifiants=None, identifiants_adjacents=None):
    """Utility function to create Gradio input fields based on the processor's key characteristics."""
    inputs = []
    for key, characteristic in processor.key_characteristics.items():
        # print(key, characteristic, type(characteristic))
        if type(characteristic) == str and characteristic == "any":
            if not(key in ['parois', 'vitrages', 'ponts_thermiques', 'installations']):
                if key == "identifiant":
                    input_field = gr.Dropdown(choices=identifiants, label=key)
                elif key == "identifiant_adjacents":
                    input_field = gr.Dropdown(choices=identifiants_adjacents, label=key,multiselect=True)
                elif "is_" in key:
                    input_field = gr.Radio([True, False], label=key, value=False)
                else:
                    input_field = gr.Textbox(label=key, placeholder="Enter text")
        elif type(characteristic) == str and characteristic == "float":
            input_field = gr.Number(label=key)
        elif isinstance(characteristic, dict) and 'min' in characteristic and 'max' in characteristic:
            input_field = gr.Slider(minimum=0, maximum=characteristic['max'], label=key)
        elif isinstance(characteristic, list) or isinstance(characteristic, np.ndarray):
            char = [elt for elt in characteristic if ((elt is not None) and (elt != "NULL"))] + [None]
            input_field = gr.Dropdown(choices=char, label=key, value=None)
        else:
            input_field = gr.Text(label="Unsupported type for " + key)
        inputs.append(input_field)
    return inputs


murs=[f"mur_{i}" for i in range(6)]
ph=[f"plancher_haut_{i}" for i in range(2)]
pb=[f"plancher_bas_{i}" for i in range(2)]
ouvrants=[f"vitrage_{i}" for i in range(10)]
# verandas=[f"veranda{i}" for i in range(2)]
ponts_thermiques=[f"pont_thermique_saisie_{i}" for i in range(20)]
ecs=[f"ecs_{i}" for i in range(3)]
clims=[f"clim_{i}" for i in range(1)]
chauffages=[f"chauffage_{i}" for i in range(3)]

green= '#519351'
black= '#000000'
blue= '#4E8FB7'
gold= '#CEAD09'
lightGreen= '#EFF5EE'
white= "#FFFFFF"
gray="#F5F5F5"

theme=gr.themes.Glass(
    primary_hue="yellow",
    secondary_hue="yellow",
    spacing_size="sm",
    text_size='lg',
    radius_size='sm',
)

theme.set(
    body_background_fill=green,
    background_fill_primary=green,
    body_text_size='text_lg',
    block_label_text_size='text_lg',
    body_text_color=gold,
    body_text_color_subdued=white,
    body_text_weight='800',
    block_background_fill=green,
    input_background_fill=gray,
)

with gr.Blocks(theme=theme) as demo:
    with gr.Tab(label="General Informations"):
        general_inputs = create_input_fields(dpe)
        gr.Row(*general_inputs)  # Use * to unpack the list of inputs

    with gr.Tab(label="Parois"):
        for j in range(10):
            with gr.Accordion("Paroi " + str(j), open=False):
                parois_inputs = create_input_fields(dpe.parois_processor, identifiants=murs+ph+pb, identifiants_adjacents=murs+ph+pb+ouvrants)
                gr.Row(*parois_inputs)

    with gr.Tab(label="Ouvrants"):
        for j in range(10):
            with gr.Accordion("Ouvrant " + str(j), open=False):
                ouvrants_inputs = create_input_fields(dpe.vitrage_processor, identifiants=ouvrants)
                gr.Row(*ouvrants_inputs)
    
    with gr.Tab(label="Ponts thermiques"):
        for j in range(10):
            with gr.Accordion("Pont thermique " + str(j), open=False):
                ponts_thermiques_inputs = create_input_fields(dpe.pont_thermique_processor, identifiants=ponts_thermiques)
                gr.Row(*ponts_thermiques_inputs)

    with gr.Tab(label="Installations ECS"):
        for j in range(3):
            with gr.Accordion("Installation ECS " + str(j), open=False):
                ecs_inputs = create_input_fields(dpe.ecs_processor, identifiants=ecs)
                gr.Row(*ecs_inputs)

    with gr.Tab(label="Installations Climatisation"):
        for j in range(1):
            with gr.Accordion("Installation Climatisation " + str(j), open=False):
                clim_inputs = create_input_fields(dpe.clim_processor, identifiants=clims)
                gr.Row(*clim_inputs)
    
    with gr.Tab(label="Installations Chauffage"):
        for j in range(3):
            with gr.Accordion("Installation Chauffage " + str(j), open=False):
                chauffage_inputs = create_input_fields(dpe.chauffage_processor, identifiants=chauffages)
                gr.Row(*chauffage_inputs)

    with gr.Tab(label="Results"):
        if gr.Button("Calculate"):
            gr.Markdown('Results will be here')
        

demo.queue(20).launch(share=True, debug=False)
```
